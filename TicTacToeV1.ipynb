```python
#version 1 - no grid
#version 2 - basic with grid
#version 3 - user names scoreboards & wasd keys
import os    
import time    
    
board = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']    
player = 1    
   
#game outcome flags    
Win = 1    
Draw = -1    
Running = 0    
Stop = 1    
#------------------------------------------    
Game = Running    
Mark = 'X'    
   
    
   
#function checking if position is available  
def CheckPosition(x):    
    if(board[x] == ' '):    
        return True    
    else:    
        return False    
   
#function checking if there is a winning combination present    
def CheckWin():    
    global Game    
    #Horizontal winning condition    
    if(board[1] == board[2] and board[2] == board[3] and board[1] != ' '):    
        Game = Win    
    elif(board[4] == board[5] and board[5] == board[6] and board[4] != ' '):    
        Game = Win    
    elif(board[7] == board[8] and board[8] == board[9] and board[7] != ' '):    
        Game = Win    
    #Vertical Winning Condition    
    elif(board[1] == board[4] and board[4] == board[7] and board[1] != ' '):    
        Game = Win    
    elif(board[2] == board[5] and board[5] == board[8] and board[2] != ' '):    
        Game = Win    
    elif(board[3] == board[6] and board[6] == board[9] and board[3] != ' '):    
        Game=Win    
    #Diagonal Winning Condition    
    elif(board[1] == board[5] and board[5] == board[9] and board[5] != ' '):    
        Game = Win    
    elif(board[3] == board[5] and board[5] == board[7] and board[5] != ' '):    
        Game=Win    
    #Match Tie or Draw Condition    
    elif(board[1]!=' ' and board[2]!=' ' and board[3]!=' ' and board[4]!=' ' and board[5]!=' ' and board[6]!=' ' and board[7]!=' ' and board[8]!=' ' and board[9]!=' '):    
        Game=Draw    
    else:            
        Game=Running    
    
    
print("Player 1 [X] --- Player 2 [O]\n")    
print()    
print()    
time.sleep(3)    
while(Game == Running):    
    os.system('cls')        
    if(player % 2 != 0):    
        print("Player 1's turn")    
        Mark = 'X'    
    else:    
        print("Player 2's turn")    
        Mark = 'O'    
    choice = int(input("Enter a number ranging from 1 to 9 in correlation to grid choosing where you want to mark : "))    
    if(CheckPosition(choice)):    
        board[choice] = Mark    
        player+=1    
        CheckWin()    
    
os.system('cls')      
if(Game==Draw):    
    print("its a draw!")    
elif(Game==Win):    
    player-=1    
    if(player%2!=0):    
        print("player 1 wins.")    
    else:    
        print("player 2 wins.")    
```
